```python
def pay(x):
    if x < 45000:
        return '<45K'
    elif x < 80000:
        return '<80K'
    elif x < 200000:
        return '<200K'
    else:
        return '>200K'
pay_index = [['<45', '<80k', '<200k', '>200k']]
df['pay'] = df['salary_in_usd'].apply(pay)
pay_percent = df['pay'].value_counts(normalize = True).mul(100).round(1).astype(str)+'%'
pay_percent
```

```python
df_c = df.corr()
one_like = np.ones_like(df_c, dtype = bool)
mask1 = np.triu(one_like)
print(df_c)

sns.heatmap(df_c, annot = True, fmt = ".2f", cmap = 'Greens')
```
